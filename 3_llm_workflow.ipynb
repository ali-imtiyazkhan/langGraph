```python
from dotenv import load_dotenv
from langgraph.graph import StateGraph,START, END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model="gemini-flash-latest", temperature=0.2)
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def ask_question(state: LLMState) -> LLMState:
    question = state["question"]

    # form a prompt for the model
    prompt = f"Answer the following question: {question}"

    # ask the question here

    answer = model.invoke(prompt).content

    # return the updated state
    state["answer"] = answer
    
    # return the answer to the graph
    return state
```

```python
# create a graph
graph = StateGraph(LLMState)

# create the nodes 

graph.add_node("ask_question", ask_question)

# add your edges here
graph.add_edge(START, "ask_question")
graph.add_edge("ask_question", END)

# compile the graph
workflow = graph.compile()
```

```python
initial_state = LLMState(question="What is the capital of France?")

final_state = workflow.invoke(initial_state)

print(final_state["answer"])
```
